```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/cars.csv"
df_cars = pd.read_csv(link)

df_cars.head()
```

```python
df_continent = pd.get_dummies(df_cars["continent"],dtype='int')
df_continent
```

```python
df_cars2 = pd.concat([df_cars, df_continent], axis=1)
df_cars2
```

```python
# Create a copy of df_cars2 with the 'Continent' column
df_cars3 = df_cars2.copy()

# Drop the 'Continent' column in df_cars3
df_cars3.drop('continent', axis=1, inplace=True)
```

```python
df_sample = df_cars2.sample(n=20, random_state=10)
sns.pairplot(df_sample)
plt.show()
```

```python
import math
num_columns = len(df_cars2.columns)
num_rows = math.ceil(num_columns / 3)
plt.figure(figsize=(15, 5 * num_rows))

for i, column in enumerate(df_cars2.columns, 1):
    plt.subplot(num_rows, 3, i)
    sns.histplot(df_cars2[column], kde=True)
    plt.title(f'Distribution de {column}')
    plt.xlabel(column)
    plt.ylabel('FrÃ©quence')
plt.tight_layout()
plt.show()
```

```python
# CrÃ©er une distribution avec Seaborn
sns.set(style="whitegrid")  # Configuration du style de la grille

# Utiliser la fonction distplot pour crÃ©er la distribution
sns.distplot(df_cars2['mpg'], kde=False, bins=20, color='skyblue')

# Ajouter des labels et un titre
plt.xlabel('Miles per Gallon (mpg)')
plt.title('Distribution of Miles per Gallon')

# Afficher le graphique
plt.show()
```
