```python
df_split = df['production_companies_country'].str.split(',', expand=True)
df_split.columns = ['production_companies_country1','production_companies_country2','production_companies_country3','production_companies_country4','production_companies_country5','production_companies_country6','production_companies_country7','production_companies_country8','production_companies_country9','production_companies_country10','production_companies_country11','production_companies_country12','production_companies_country13','production_companies_country14','production_companies_country15','production_companies_country16','production_companies_country17','production_companies_country18','production_companies_country19','production_companies_country20','production_companies_country21','production_companies_country22','production_companies_country23','production_companies_country24','production_companies_country25','production_companies_country26','production_companies_country27','production_companies_country28','production_companies_country29','production_companies_country30']
df_split_concat = pd.concat([df, df_split], axis=1)
df = df_split_concat
df
```

```python
df = df[df.index != 'XC']
df
```

```python
# Remplacement de la valeur '\N' par 'Nan' qui est plus conventionnelle
df.replace("['']", pd.NA, inplace=True)
df.replace("[]", pd.NA, inplace=True)
df.replace("[''", pd.NA, inplace=True)
df.replace("nan", pd.NA, inplace=True)
df.dropna(axis=0, how='any', inplace=True, subset="production_companies_country1")

df["production_companies_country1"]
```

```python
nombre_valeurs_nulles = df["production_companies_country1"].dropna(inplace=True)
print(nombre_valeurs_nulles)
```

```python
lang = df["production_companies_country1"].unique()
language = []
for i in lang :
    prefixe_sup = i.replace("['","")
    suffixe_sup = prefixe_sup.replace("']","")
    suffixe_sup = suffixe_sup.replace("'","")
    language.append(suffixe_sup)
    
print(language)
```

```python
#df.replace("['US'", "['US']", inplace=True)
#df.replace("['FR'", "['FR']", inplace=True)
#df.replace("['GB'", "['GB']", inplace=True)
import requests
import json

for i in language:
    df.replace("['"+i+"'", i, inplace=True)
    df.replace("['"+i+"']",i, inplace=True)

#Téléchargement de la liste des country codes / noms des pays en français
r = requests.get('https://gist.githubusercontent.com/lneveu/cdb444b0e609ed81d3ad1f5907cda6f8/raw/8ad6e298e55284b074dcd716da2d3b25904c29f8/iso-3166_country_french.json')
country_code = r.json()

#Adding missing country code
country_code['SU'] = 'URSS'

#Remplacement des country codes par le nom des pays en entier
for key,country in country_code.items():
    for lang in language :
        if lang == key :
            df.replace(lang, country, inplace=True)

df_values = df["production_companies_country1"].value_counts().head(15)
df_values       

```
